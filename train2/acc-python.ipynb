```python
from ultralytics import YOLO
import cv2
import os
import matplotlib.pyplot as plt
model_path='/Users/vaishalilalit/Downloads//train/weights/best.pt'
input_dir='/Users/vaishalilalit/Downloads/train/mychtest/test'
output_dir='/Users/vaishalilalit/Downloads/train/mychtest/mask'
for j in os.listdir(input_dir):
    image_path = os.path.join(input_dir, j)
    image=cv2.imread(image_path)
#     plt.imshow( image)
    H,W,_=image.shape
    model=YOLO(model_path)
    results=model(image)
    base_filename = os.path.splitext(j)[0]  # Extracting the base filename without extension

    for i, result in enumerate(results):
        if result.masks is not None and hasattr(result.masks, 'data'):
            for k, mask in enumerate(result.masks.data):
                mask = mask.numpy() * 255
                mask = cv2.resize(mask, (W, H)).astype('uint8')

                # Save the mask with the original filename and an index
                mask_filename = '{}.jpg'.format(base_filename)
                mask_filepath = os.path.join(output_dir, mask_filename)
                cv2.imwrite(mask_filepath, mask)
                print(f"Saved mask at: {mask_filepath}")
        else:
            print(f"No masks found for: {os.path.join(output_dir, j)}")
```

```python
import json
import numpy as np
import os
from PIL import Image

# List of supported image file extensions
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp', '.gif']

all_dict = {}

folder_path = '/Users/vaishalilalit/Downloads/train/mychtest/mask'

for file_name in os.listdir(folder_path):
    if any(file_name.lower().endswith(ext) for ext in image_extensions):
        temp_dict = {}
        image_path = os.path.join(folder_path, file_name)
        image = Image.open(image_path)
        
        # Perform image processing or other operations here
        new_array = np.array(image) / 255
        new_array = new_array.astype(int)
        
        temp_dict["counts"] = mask_to_rle(new_array)
        temp_dict["height"] = image.size[1]
        temp_dict["width"] = image.size[0]
        all_dict[file_name] = temp_dict

# Convert dictionary to JSON string
json_obj = json.dumps(all_dict)

# Write JSON string to a file
with open('/Users/vaishalilalit/Downloads/train/mychtest/results.json', 'w') as f:
    f.write(json_obj)
```

```python
def mask_to_rle(mask: np.ndarray):
    """
    Convert a binary mask to RLE format.
    :param mask: numpy array, 1 - mask, 0 - background
    :return: RLE array
    """
    pixels = mask.T.flatten()
    pixels = np.concatenate([[0], pixels, [0]])
    runs = np.where(pixels[1:] != pixels[:-1])[0] + 1
    runs[1::2] -= runs[::2]
    return [int(x) for x in runs]
```
